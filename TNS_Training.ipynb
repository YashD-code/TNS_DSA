STACK

```python
class Stack:
  def __init__(self):
    self.stack=[]
  def push(self,item):
    self.stack+=[item]
  def pop(self):
    if self.stack:
      top_element=self.stack[-1]
      self.stack=self.stack[:-1]
      return top_element
    else:
      return "Stack Underflow"
  def isEmpty(self):
    return len(self.stack)==0
  def peek(self):
    if self.stack:
      return self.stack[-1]
    else:
      return "Stack Underflow"
s=Stack()
s.push(10)
s.push(20)
s.push(30)
print(s.pop())
print(s.peek())
print(s.isEmpty())
```

Balanced paranthesis

```python
def isBalanced(expression):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for char in expression:
        if char in '({[':
            stack.append(char)
        elif char in ')}]':
            if not stack or stack[-1] != pairs[char]:
                return "Not Balanced"
            stack.pop()

    return "Balanced" if not stack else "Not Balanced"

# Test cases
print(isBalanced("{[()]}"))
print(isBalanced("{[()])}"))
```

sort a stack (using another stack)

```python
def sortstack(ogstack):
  tempstack=[]
  while ogstack:
    temp=ogstack.pop()
    while tempstack and tempstack[-1]>temp:
      ogstack.append(tempstack.pop())
    tempstack.append(temp)
  return tempstack
arr=[1,8,2,3,67,7,8,9,223,4]
ans=sortstack(arr)
print(ans)
```

QUEUE

```python
class Queue:
  def __init__(self):
    self.queue=[]
  def enqueue(self,item):
    self.queue+=[item]
  def dequeue(self):
    if self.queue:
      topElement=self.queue[0]
      self.queue=self.queue[1:]
    else:
      return "Queue Underflow"
    return topElement
  def peek(self):
    if self.queue:
      return self.queue[0]
    else:
      return "queue underflow"
  def isEmpty(self):
    return len(self.queue)==0
  def size(self):
    return len(self.queue)
q=Queue()
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
print(q.dequeue())
print(q.peek())
print(q.isEmpty())
print(q.size())
```

Generate Binary Numbers 1 to N

```python
from collections import deque
def binaryNumber(n):
  q=deque()
  q.append("1")
  ans=""
  for i in range(n):
    temp=q.popleft()
    q.append(temp+"0")
    q.append(temp+"1")
    ans+=temp+" "
  print(ans)

n=int(input("Enter the N value:"))
a=binaryNumber(n)
print(a)
```
